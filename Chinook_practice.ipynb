```python
##Select all the customers from every country other than the USA
query = """
select CustomerId, FirstName, LastName, Country
from Customers
where not Country = "USA";
"""
pd.read_sql_query(query,conn)
```

```python
#select name, invoice, invoice date from usa customers
query = """
select c.FirstName, c.LastName, i.InvoiceId, i.InvoiceDate, i.BillingCountry
from Customers as c, Invoices as i
where c.Country = "USA" and
c.CustomerId = i.CustomerId;
"""
pd.read_sql_query(query,conn)
```

```python
#show invoices associated with each sales agent 
query = """
select e.FirstName, e.LastName, i.InvoiceId, i.CustomerId, i.InvoiceDate, i.BillingAddress, i.BillingCountry, i.BillingPostalCode, i.Total
from Customers as c, Invoices as i
on c.CustomerId = i.CustomerId
join Employees as e
on e.EmployeeId = c.SupportRepId
order by e.EmployeeId
"""
pd.read_sql_query(query,conn)
```

```python
#query that shows invoice total, customer name, country, sale agent for all invoices and customers
query = """
select e.FirstName as "employee first name", e.LastName "employee last name", c.FirstName "customer first name", c.LastName "customer last name", c.Country, i.Total
from Employees as e
join Customers as c on e.EmployeeId = c.SupportRepId
join Invoices as i on c.CustomerId = i.CustomerId
"""
pd.read_sql_query(query,conn)
```

```python
#invoices between 2009 and 2011 and respective total sales for those years
query = """
select count(i.InvoiceId), sum(i.Total)
from Invoices as i
where i.InvoiceDate between datetime("2009-01-01 00:00:00") and datetime("2009-12-31 00:00:00");
"""
pd.read_sql_query(query,conn)
```

```python
#number of line items for each invoice
query = """
select InvoiceId, count(InvoiceLineId)
from Invoice_Items
group by InvoiceId
"""
pd.read_sql_query(query,conn)
```

```python
#includes the track name with each invoice line item
query = """
select i.*, t.Name
from Invoice_Items as i, Tracks as t
on i.TrackId = t.TrackId
"""
pd.read_sql_query(query,conn)
```

```python
#invoices per country
query = """
select BillingCountry, count(BillingCountry) as "# of invoices"
from Invoices
group by BillingCountry
"""
pd.read_sql_query(query,conn)
```

```python
#shows the total number of invoices per country
query = """
select *, count(TrackId) as "# of tracks"
from Playlist_Track, PlayLists
on Playlist_Track.PlaylistId = Playlists.PlaylistId
group by Playlists.PlaylistId
"""
pd.read_sql_query(query,conn)
```

```python
#provide a query that shows all the tracks, but displays no ids. include album name, media type, and genre
query = """
select t.Name as "track", t.Composer, t.Milliseconds, t.Bytes, t.UnitPrice, a.Title as "album", g.Name as "genre", m.Name as "media type"
from Tracks as t
join Albums as a on a.AlbumId = t.AlbumId
join Genres as g on g.GenreId = t.GenreId
join Media_Types as m on m.MediaTypeId = t.MediaTypeId
"""
pd.read_sql_query(query,conn)
```

```python
#shows all invoices but includes the # of invoice line items
query = """
select Invoices.*, count(Invoice_Items.InvoiceLineId) as "# of line items"
from Invoices, Invoice_Items
on Invoices.InvoiceId = Invoice_Items.InvoiceId
group by Invoices.InvoiceId
"""
pd.read_sql_query(query,conn)
```

```python
#total sales made by each sales agent
query = """
select e.*, count(i.InvoiceId) as "total number of sales"
from Employees as e
join Customers as c on e.EmployeeId = c.SupportRepId
join Invoices as i on i.CustomerId = c.CustomerId
group by e.EmployeeId
"""
pd.read_sql_query(query,conn)
```

```python
#which sales agents made the most in sales in 2009
query = """
select *, max(Total) from
(select e.*, sum(Total) as "Total"
from Employees as e
join Customers as c on e.EmployeeId = c.SupportRepId
join Invoices as i on i.CustomerId = c.CustomerId
where i.InvoiceDate between datetime("2009-01-01 00:00:00") and datetime("2009-12-31 00:00:00")
group by e.EmployeeId)
"""
pd.read_sql_query(query,conn)
```

```python
#which sales agent made the most in sales i 2010
query = """
select *, max(Total) from
(select e.*, sum(Total) as "total"
from Employees as e
join Customers as c on e.EmployeeId = c.SupportRepId
join Invoices as i on i.CustomerId = c.CustomerId
where i.InvoiceDate between "2010-01-00" and "2010-12-31"
group by e.EmployeeId)
"""
pd.read_sql_query(query,conn)
```

```python
#which sales agent made the most sales overall
query = """
select *, max(Total) from
(select e.*, sum(Total) as "Total"
from Employees as e
join Customers as c on e.EmployeeId = c.SupportRepId
join Invoices as i on i.CustomerId = c.CustomerId
group by e.EmployeeId)
"""
pd.read_sql_query(query,conn)
```

```python
#shows the # of customers assigned to each sale agent
query = """
select e.*, count(c.CustomerId) as "total customers"
from Employees as e
join Customers as c on e.EmployeeId = c.SupportRepId
group by e.EmployeeId
"""
pd.read_sql_query(query,conn)
```

```python
#provide a query that shows the total sales per country. Which country's customers spent the most
query = """
select i.BillingCountry, sum(Total) as "totalsales"
from Invoices as i
group by BillingCountry
order by totalsales desc
"""
pd.read_sql_query(query,conn)
```

```python
#provide a query that shows the most purchased track of 2013
query = """
select *, count(t.TrackId) as count
from Invoice_Items as il
join Invoices as i on i.InvoiceId = il.InvoiceId
join Tracks as t on t.TrackId = il.TrackId
where i.InvoiceDate between "2013-01-01" and "2013-12-31"
group by t.TrackId
order by count desc
"""
pd.read_sql_query(query,conn)
```

```python
#top 3 selling artists
query = """
select a.Name as "Artist", sum(it.UnitPrice) as "Total sold"
from Invoice_Items as it, Tracks as t, Albums as al, Artists as a
where it.TrackId = t.TrackId
  and al.AlbumId = t.AlbumId
  and a.ArtistId = al.ArtistId
group by a.Name
order by count(a.ArtistId) desc
limit 3;
"""
pd.read_sql_query(query,conn)
```

```python
#top media types
query = """
select m.Name, count(m.MediaTypeId) as Amount
from Media_Types as m
inner join Tracks as t on m.MediaTypeId = t.MediaTypeId
group by m.MediaTypeId
order by Amount desc;
"""
pd.read_sql_query(query,conn) 
```

```python
#top 10 rock bands
query = """
select a.Name ArtistName, count(t.TrackId) as TrackCount from Tracks as t
inner join Albums as al on t.AlbumId = al.AlbumId
inner join Artists as a on al.ArtistId = a.ArtistId
inner join Genres as g on t.GenreId = g.GenreId
where g.GenreId = 1
group by al.ArtistId
order by TrackCount desc limit 10
"""
pd.read_sql_query(query,conn) 
```

```python
#best customer
query = """
select (c.FirstName || " " || c.LastName) as CustomerName, round(sum(i.Total), 2) as AmountSpent from Invoices as i
inner join Customers as c on i.CustomerId = c.CustomerId
inner join Invoice_Items as it on i.InvoiceId = it.InvoiceId
group by c.FirstName, c.LastName
order by AmountSpent desc
limit 1;
"""
pd.read_sql_query(query,conn) 
```

```python
#States with the most invoices
query = """
select BillingState, count(BillingState) as invoices
from Invoices
where BillingCountry = "USA"
group by BillingState
order by invoices desc
"""
pd.read_sql_query(query,conn)
```

```python
#Number of classical albums
query = """
select count(al.AlbumId) as "# Albums"
from Albums as al
join Tracks as t on al.AlbumId = t.AlbumId
join Genres as g on g.GenreId = t.GenreId
where g.Name = "Classical"
"""
pd.read_sql_query(query,conn)
```

```python
#Total sales of classical tracks
query = """
select count(it.TrackId) as totalsales
from Invoice_Items as it
join Tracks as t on it.TrackId = t.TrackId
join Genres as g on t.GenreId = t.GenreId
where g.Name = "Classical"
"""
pd.read_sql_query(query,conn)
```

```python
#Customer that spent the most on classical tracks
query = """
select (c.FirstName || " " || c.LastName) as CustomerName, round(sum(i.Total), 2) as AmountSpent from Invoices as i
inner join Customers as c on i.CustomerId = c.CustomerId
inner join Invoice_Items as it on i.InvoiceId = it.InvoiceId
join Tracks as t on it.TrackId = t.TrackId
join Genres as g on t.GenreId = g.GenreId
where g.Name = "Classical"
group by c.FirstName, c.LastName
order by AmountSpent desc
limit 1;
"""
pd.read_sql_query(query,conn) 
```

```python
#Album total sales
query = """
select al.Title, count(i.Total) as totalsales
from Albums as al
join Tracks as t on al.AlbumId = t.AlbumId
join Invoice_Items as it on t.TrackId = it.TrackId
join Invoices as i on it.InvoiceId  = i.InvoiceId
group by Title
order by totalsales desc limit 10
"""
pd.read_sql_query(query,conn) 
```

```python
#Employees with the most customers
query = """
select e.FirstName as employee, count(c.CustomerId) as customers
from Customers as c
join Employees as e on c.SupportRepId = e.EmployeeId
group by employee
order by customers desc
"""
pd.read_sql_query(query,conn)
```

```python
#Total sales of different genres
query = """
select g.Name genre, count(*) as totalsales
from Genres as g
inner join Tracks as t on g.GenreId = t.GenreId
inner join Invoice_Items as it on t.TrackId = it.TrackId
inner join Invoices as i on it.InvoiceId = i.InvoiceId
group by g.Name
order by totalsales desc limit 10
"""
pd.read_sql_query(query,conn) 
```

```python
#Total sales of different artists
query = """
select a.Name artist, count(*) as totalalbums
from Artists as a
inner join Albums as al on a.ArtistId = al.ArtistId
group by a.Name
order by totalalbums desc limit 10
"""
pd.read_sql_query(query,conn) 
```

```python
#Songs on back to black in alphabetical order
query = """
select t.Name as song, al.Title
from Tracks as t
join Albums as al on t.AlbumId = al.AlbumId
where al.Title = "Back to Black"
order by t.Name
"""
pd.read_sql_query(query,conn) 
```

```python
#Artists without any albums
query = """
select a.Name, al.Title, a.ArtistId
from Artists a
left join Albums as al on a.ArtistId = al.ArtistId
where al.Title is NULL
order by a.ArtistId asc
"""
pd.read_sql_query(query,conn) 
```

```python
#Album length by milliseconds
query = """
select al.Title, t.Milliseconds
from Albums as al
inner join Tracks as t on al.AlbumId = t.AlbumId
group by Title
order by t.Milliseconds desc limit 10
"""
pd.read_sql_query(query,conn) 
```

```python
#Cities outside of the USA with the most sales
query = """
select BillingCity, count(BillingCity) as invoices
from Invoices
where not BillingCountry = "USA"
group by BillingCity
order by invoices desc
"""
pd.read_sql_query(query,conn)
```
